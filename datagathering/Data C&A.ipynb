```python
import json

raw_data = {}
with open("data/c_n_a_raw.json", "r") as f:
    raw_data = json.loads(f.read())
```

```python
raw_data.keys()
```

```python
raw_data["suppliers"]
```

```python
processed_data = []

for d in raw_data["suppliers"]:
    company_type = "NONE"
    if d["tier1"] == 1:
        company_type = "FACTORY"
    elif d["tier2"] == 1:
        company_type = "TIER 2"

    point = {
        "brand": "C&A",
        "company_type": company_type,
        "country": d["country"],
        "name": d["name"],
        "lat": d["lat"],
        "lon": d["lng"],        
    }
    
    processed_data.append(point)
```

```python
import json
with open("data/c_n_a_locations.json", "w") as f:
    data = {
        "factories": processed_data,
    }
    f.write(json.dumps(data))
```
